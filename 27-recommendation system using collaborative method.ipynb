```python
import pandas as pd
```

```python
books=pd.read_csv("C:\\Users\\user\\Downloads\\BX-Books.csv",sep=";",encoding="latin-1",error_bad_lines=False)
```

```python
users=pd.read_csv("C:\\Users\\user\\Downloads\\BX-Users.csv",sep=";",encoding="latin-1",error_bad_lines=False)
```

```python
ratings=pd.read_csv("C:\\Users\\user\\Downloads\\BX-Book-Ratings.csv",sep=";",encoding="latin-1",error_bad_lines=False)
```

```python
books.head()
```

```python
books=books[['ISBN','Book-Title','Book-Author','Year-Of-Publication','Publisher']]
```

```python
books=books.rename(columns={'Book-Title':'title','Book-Author':'author','Year-Of-Publication':'year','Publisher':'publisher'})
```

```python
users=users.rename(columns={'User-ID':'user_id',"Location":"location","Age":"age"})
```

```python
ratings=ratings.rename(columns={"User-ID":"user_id","Book-Rating":"rating"})
```

```python
#renaming the title of headings of the columns above
```

```python
books.head()
```

```python
#collaborative filtering
```

```python
#1.Take only those users who have rated atleast 200 books
#2.Take only those books which have received atleast 50 ratings from the user
```

```python
x=ratings['user_id'].value_counts()>200
x
```

```python
y=x[x].index #userid who has rated more than 200 books
y
```

```python
ratings[ratings['user_id'].isin(y)] #extracting those user ids exactly in y
```

```python
rating_with_books=ratings.merge(books,on='ISBN')
rating_with_books #rating is also added to book data
```

```python
number_rating=rating_with_books.groupby('title')['rating'].count().reset_index()
number_rating
#we are grouping book based on title and  aggregate based on rating
```

```python
number_of_ratings=number_rating.rename(columns={'rating':'number_of_ratings'},inplace=True)
```

```python
rating_with_books #merge this with nu.of books
```

```python
final_rating=rating_with_books.merge(number_rating,on="title")
final_rating #to get which user has rated which movie
```

```python
final_rating=final_rating[final_rating['number_of_ratings']>50]
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating
```

```python
#pivot table
```

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='title',values='rating')
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
book_pivot
```

```python
#modelling K-nearest neighbour algorithm
```

```python
#spars matrix will use if you have most of values as zero
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse=csr_matrix(book_pivot)
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance,suggestions=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1)) #237 is random,you can give any number to be there the targeted user
```

```python
distance
```
