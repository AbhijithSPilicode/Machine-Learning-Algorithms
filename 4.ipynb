```python
classes=["python","R","ML","AI","DS"]
class1_students=[30,10,20,25,10]
colors=["r","b","g","c","y"]
plt.pie(class1_students,labels=classes,colors=colors)
```

```python
plt.pie(class1_students,labels=classes,colors=colors,autopct="%0.2f%%")
```

```python
plt.pie(class1_students,labels=classes,colors=colors,autopct="%0.2f%%",explode=[0.2,0,0,0,0])
```

```python
plt.pie(class1_students,labels=classes,colors=colors,autopct="%0.2f%%",explode=[0.2,0.1,0.1,0,0.2])
```
